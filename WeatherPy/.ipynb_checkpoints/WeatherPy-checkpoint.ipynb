# WeatherPy

```python
import os
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import numpy as np
import datetime
from scipy.stats import linregress
from citipy import citipy
from keys import owkey
from keys import gkey
```

# Generate Cities

```python
latrange = (-90, 90)
longrange = (-180, 180)
latlongs = []
cities = []
lats = np.random.uniform(low=-90, high=90, size=1500)
longs = np.random.uniform(low=-180, high=180, size=1500)
latlongs = zip(lats, longs)
for latlong in latlongs:
    city = citipy.nearest_city(latlong[0], latlong[1]).city_name
    if city not in cities:
        cities.append(city)
print(cities)
```

# API Setup

```python
burl = "http://api.openweathermap.org/data/2.5/weather?"
weather = pd.DataFrame({"City": cities})
weather["Country"] = ""
weather["Date"] = ""
weather["Latitude"] = ""
weather["Longitude"] = ""
weather["Cloudiness"] = ""
weather["Humidity"] = ""
weather["Max Temp"] = ""
weather["Wind Speed"] = ""
for index, row in weather.iterrows():
    city = row["City"]
    qurl = burl + "appid=" + owkey + "&q=" + city
    weatherresp = requests.get(qurl).json()
    
    try:
        weather.loc[index, "Country"] = weatherresp["sys"]["country"]
        weather.loc[index, "Date"] = datetime.datetime.fromtimestamp(weatherresp["dt"])
        weather.loc[index, "Latitude"] = weatherresp["coord"]["lat"]
        weather.loc[index, "Longitude"] = weatherresp["coord"]["lon"]
        weather.loc[index, "Cloudiness"] = weatherresp["clouds"]["all"]
        weather.loc[index, "Humidity"] = weatherresp["main"]["humidity"]
        weather.loc[index, "Max Temp"] = 1.8 * (weatherresp["main"]["temp_max"] - 273) + 32
        weather.loc[index, "Wind Speed"] = weatherresp["wind"]["speed"]
    except:
        print(f"City not found, skipping {city}.")
    
    
    
```

# Linear Regression

```python
north = weather.loc[pd.to_numeric(weather["Latitude"]).astype(float) > 0, :]
south = weather.loc[pd.to_numeric(weather["Latitude"]).astype(float) < 0, :]
nclouds = pd.to_numeric(north["Cloudiness"]).astype(float)
nhumid = pd.to_numeric(north["Humidity"]).astype(float)
nlats = pd.to_numeric(north["Latitude"]).astype(float)
nlons = pd.to_numeric(north["Longitude"]).astype(float)
nmaxs = pd.to_numeric(north["Max Temp"]).astype(float)
nspeed = pd.to_numeric(north["Wind Speed"]).astype(float)
sclouds = pd.to_numeric(south["Cloudiness"]).astype(float)
shumid = pd.to_numeric(south["Humidity"]).astype(float)
slats = pd.to_numeric(south["Latitude"]).astype(float)
slons = pd.to_numeric(south["Longitude"]).astype(float)
smaxs = pd.to_numeric(south["Max Temp"]).astype(float)
sspeed = pd.to_numeric(south["Wind Speed"]).astype(float)
```

## Northern Hemisphere - Temperature (F) vs. Latitude

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(nlats, nmaxs)
rvals = nlats * slope + intercept
lineeq = "y = " + str(round(slope, 2))+ "x + " + str(round(intercept,2))
plt.scatter(nlats, nmaxs, facecolors='b', edgecolors='r', alpha = 0.5)
plt.plot(nlats, rvals, "r-")
plt.annotate(lineeq, (10,40), fontsize=15, color='r')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("Northern Hemisphere - Temperature (F) vs. Latitude")
print("The regression shows a negative correlation, which means the further from the equator, the lower the temperature.")
plt.show()
```

## Southern Hemisphere - Temperature (F) vs. Latitude

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(slats, smaxs)
rvals = slats * slope + intercept
lineeq = "y = " + str(round(slope, 2))+ "x + " + str(round(intercept,2))
plt.scatter(slats, smaxs, facecolors='b', edgecolors='r', alpha = 0.5)
plt.plot(slats, rvals, "r-")
plt.annotate(lineeq, (-50,90), fontsize=15, color='r')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("Southern Hemisphere - Temperature (F) vs. Latitude")
print("The regression shows a positive correlation, which means the closer to the equator, the higher the temperature.")
plt.show()
```
